## `for`-Loops

### Beispiel
Wir haben alle Fakultäten der OTH Regensburg in einer Liste gespeichert und die zugehörigen Zahlen an Studenten in einer anderen Liste.

Unsere Aufgabe in diesem Beispiel ist es
* den Fakultätsnamen und die zugehörige Studentenanzahl in einem passenden Satz ausgeben zu lassen
* und nach dem `for`-Loop die Gesamtzahl aller Studenten in einem geeigneten Satz.

```python
list_of_number_of_students = [793, 699, 937, 2014, 1275, 1876, 1683, 1793]
oth_faculties = ['Angewandte Natur- und Kulturwissenschaften',
                 'Architektur',
                 'Bauingenieurwesen',
                 'Betriebswirtschaft',
                 'Elektro- und Informationstechnik',
                 'Informatik und Mathematik',
                 'Maschinenbau',
                 'Angewandte Sozial- und Gesundheitswissenschaften']
```

#### Lösung

```python
number_of_all_students = 0
for index_oth_faculty in range(len(oth_faculties)):
    print(f"Die Fakultät {oth_faculties[index_oth_faculty]} hat {list_of_number_of_students[index_oth_faculty]} Studierende.")
    number_of_all_students = number_of_all_students + list_of_number_of_students[index_oth_faculty]
print(f"Die Anzahl aller Studenten an der OTH Regensburg ist {number_of_all_students}.")
```

### Beispiel: Handling some Sales Data
Wir haben die Absatzzahlen verschiedener Kunden quartalsweise für drei Jahre in einem Dictionary gespeichert.

```python
# Import
import numpy as np

# Create data
sales_dict = {
    'Baldi': {
        '2017': (np.round(np.random.rand(52,) * 1e6)).tolist(),
        '2018': (np.round(np.random.rand(52,) * 1e6)).tolist(),
        '2019': (np.round(np.random.rand(52,) * 1e6)).tolist()    
    },
    'Friedl': {
        '2017': (np.round(np.random.rand(52,) * 1e6)).tolist(),
        '2018': (np.round(np.random.rand(52,) * 1e6)).tolist(),
        '2019': (np.round(np.random.rand(52,) * 1e6)).tolist()    
    },
    'Pfenny': {
        '2017': (np.round(np.random.rand(52,) * 1e6)).tolist(),
        '2018': (np.round(np.random.rand(52,) * 1e6)).tolist(),
        '2019': (np.round(np.random.rand(52,) * 1e6)).tolist()    
    }
}
```

Wir wollen nun mittels `for`-Loops und dem Zählen innerhalb dieser ermitteln
1. wie der durchschnittliche Absatz pro Kunde pro Jahr ist
2. wie der durchschnittliche Absatz über alle Kunden und Jahre ist

Wir erinnern uns an die `keys()` Methode eines Dictionaries und die `sum()` und `len()` Methode.

Wir erarbeiten uns hier etwas Neues: sogenannte *Nested Loops*

#### Lösung

```python
average_sales = 0
running_index = 0
for customer in sales_dict.keys():
    for year in sales_dict[customer].keys():
        average_sales_per_customer = sum(sales_dict[customer][year]) / len(sales_dict[customer][year])
        print(f'Der mittlere Absatz von {customer} im Jahr {year} beträgt {average_sales_per_customer}.')
        average_sales = average_sales + average_sales_per_customer
        running_index += 1
print(f'Der mittlere Absatz über alle Kunden und Jahre beträgt {average_sales / running_index}.')
```

## `while`-Loops

### Beispiel
Wir bauen uns ein Programm, das uns eine Zahl raten lässt.

Wir erinnern uns an die `input()` Funktion. Mit ihr kann man eine User-Eingabe erfragen. Unsere Aufgabe ist es eine Variable von einem User setzen zu lassen und eine `while`-Schleife so lange ausführen zu lassen, bis wir die gleiche Zahl eingeben, die der Computer in einer anderen Variablen per Zufall gesetzt hat.

#### Lösung

```python
# Import a function that returns a random integer
import numpy as np
random_value = np.random.randint(0, 10)

# While loop that iterates until the user input is equal to random_value
user_input = -1
while random_value != user_input:
    user_input = input("Bitte eine Zahl eingeben: ")
    user_input = int(user_input)
print(f"Meine Zufallszahl ist {random_value}.")
```

## Conditions und `if`-Statements

### Beispiel
Wir wollen mittels eines for-Loops und einer `if`-Abfrage den kleinsten Wert in einer Liste finden.

#### Lösung

```python
my_list = [9, 41, 12, 3, 74, 15]
smallest_value = my_list[0]
for item in my_list:
    if item < smallest_value:
        smallest_value = item
print(smallest_value)
```

### Beispiel: Handling some Sales Data 2.0
Nun liegen uns monatliche Absatzdaten mehrerer Kunden in einem Dictionary vor. Wir wollen uns die Monate pro Kunde ausgeben lassen, die unter (absatzschwache Monate) und über (absatzstarke Monate) einer bestimmten Grenze liegen. Hierzu verändern wir das obige Beispiel durch `if`-Abfragen.

```python
# Import
import numpy as np

# Create data
sales_dict = {
    'Baldi': {
        '2017': (np.round(np.random.rand(12,) * 1e7)).tolist(),
        '2018': (np.round(np.random.rand(12,) * 1e7)).tolist(),
        '2019': (np.round(np.random.rand(12,) * 1e7)).tolist()    
    },
    'Friedl': {
        '2017': (np.round(np.random.rand(12,) * 1e7)).tolist(),
        '2018': (np.round(np.random.rand(12,) * 1e7)).tolist(),
        '2019': (np.round(np.random.rand(12,) * 1e7)).tolist()    
    },
    'Pfenny': {
        '2017': (np.round(np.random.rand(12,) * 1e7)).tolist(),
        '2018': (np.round(np.random.rand(12,) * 1e7)).tolist(),
        '2019': (np.round(np.random.rand(12,) * 1e7)).tolist()    
    }
}
```

#### Lösung

```python
threshold_for_weak_month = 1e6
threshold_for_strong_month = 9e6
for customer in sales_dict.keys():
    for year in sales_dict[customer].keys():
        count_of_months = 1
        for sales in sales_dict[customer][year]:
            if sales < threshold_for_weak_month:
                print(f"Für Kunde {customer} war der {count_of_months}. Monat des Jahres {year} ein absatzschwacher Monat.")
            elif sales > threshold_for_strong_month:
                print(f"Für Kunde {customer} war der {count_of_months}. Monat des Jahres {year} ein absatzstarker Monat.")                        
            count_of_months = count_of_months + 1
```

### Beispiel: Using List Comprehensions to filter SARS-CoV-2 data
\#listcomprehensions

Unten habe ich Ihnen angegeben, wie Sie mit pandas (wir kommen noch darauf zu sprechen) in einer Zeile eine Liste an tagesaktuellen SARS-CoV-2 Fällen aus dem Internet laden können (in diesem Beispiel handelt es sich um aktuelle Daten aus Australien). Wir nutzen nun List Comprehensions, um aus dieser Liste an Fallzahlen nur die Listenelemente in eine neue Liste `days_above_arbitrary_limit` zu schreiben, die über einen Wert von 200 liegen. Andernfalls schreiben wir eine `0` als Listenelement in diese neue Liste. Plotten Sie danach die `list_of_corona_cases_up_to_today` und die gefilterten Werte in einem Plot.

Dann googlen Sie danach, wie man Bereiche in einem Plot mit einer Farbe hinterlegt (`plt.fill_between`). Hinterlegen Sie die Tage über einen Schwellwert von 200 mit der Farbe rot, die Tage darunter mit Grün. Hierzu müssen Sie zuvor mit List Comprehensions eine Variable `is_above_arbitrary_limit` und `is_below_arbitrary_limit` erzeugen, in denen sich für Werte überhalb bzw. unterhalb des Limits `True` bzw. `False` als Eintrag befindet.

Diese Berechnungen entsprechen nicht der bei uns geläufigen 7-Tage-Inzidenz-pro-1e5-Einwohner. Um das Beispiel nicht zu komplex zu machen, beschränken wir uns hier erstmal auf diese Art der Schwellwertziehung.

</div>

#### Lösung

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Get recent SARS-CoV-2 data from some API (google for German data)
list_of_corona_cases_up_to_today = pd.read_json('https://www.theage.com.au/interactive/2020/coronavirus/data-feeder/covid-19-new-cases-json.json?v=3').values[:, 1]

# Using list comprehension: days above arbitrary limit
days_above_arbitrary_limit = [item if item > 200 else 0 for item in list_of_corona_cases_up_to_today]

# Plot curves
plt.plot(list_of_corona_cases_up_to_today, color='black')
plt.plot(days_above_arbitrary_limit, color='red')

# Highlight limit intervals
is_above_arbitrary_limit = [True if item > 200 else False for item in list_of_corona_cases_up_to_today ]
is_below_arbitrary_limit = [True if item <= 200 else False for item in list_of_corona_cases_up_to_today ]
plt.fill_between(range(len(days_above_arbitrary_limit)), y1=0, y2=max(days_above_arbitrary_limit), where=is_above_arbitrary_limit, alpha=0.2, color='red')
plt.fill_between(range(len(days_above_arbitrary_limit)), y1=0, y2=max(days_above_arbitrary_limit), where=is_below_arbitrary_limit, alpha=0.2, color='green')
```
